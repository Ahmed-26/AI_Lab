### DFS with Stack

```python
tree = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': [],
    'F': ['G'],
    'G': []
}

def dfs_stack(start, goal):
    visited = []
    stack = [start]

    while stack:
        current = stack.pop()
        if current not in visited:
            visited.append(current)
            print("Visiting:", current)
            if current == goal:
                print("Goal:", goal, "found!")
                return visited
            for child in reversed(tree[current]):
                if child not in visited:
                    stack.append(child)

    print("Goal:", goal, "not found.")
    return visited

visited_nodes = dfs_stack('A', 'C')
print("DFS Traversal:", visited_nodes)


#or

```

### BFS with Queue (using simple list)

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': ['G', 'H'],
    'E': [],
    'F': ['I', 'K'],
    'G': [],
    'H': ['L'],
    'I': [],
    'K': ['M'],
    'L': [],
    'M': []
}

def bfs_with_queue(start, goal):
    visited = []
    queue = [start] 

    while queue:
        current = queue.pop(0) 
        
        if current not in visited:
            visited.append(current)
            print("Visiting:", current)
            
            if current == goal:
                print("Goal:", goal, "found!")
                return visited
            
            # Add all children to the queue
            for child in graph[current]:
                if child not in visited and child not in queue:
                    queue.append(child) 

    print("Goal:", goal, "not found.")
    return visited

# Test the BFS with queue
print("BFS with Queue Implementation:")
visited_nodes = bfs_with_queue('A', 'K')
print("BFS Traversal:", visited_nodes)
print()

# Test with another goal
print("BFS searching for 'M':")
visited_nodes2 = bfs_with_queue('A', 'M')
print("BFS Traversal:", visited_nodes2)
print()

print("Complete BFS traversal from A:")
visited_all = bfs_with_queue('A', 'NotFound')  
print("Complete BFS order:", visited_all)
```
